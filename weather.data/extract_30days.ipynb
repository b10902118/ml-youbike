```python
import json
import os
```

```python
folder_path = "./30day"

taipei_data = []
for filename in os.listdir(folder_path):
    print(filename)
    if filename.endswith(".json"):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, "r") as f:
            file_data = f.read()  # .lstrip().rstrip()
            data = json.loads(file_data)
            taipei_data += data["cwaopendata"]["resources"]["resource"]["data"]["surfaceObs"]["location"][3]["stationObsTimes"]["stationObsTime"]
```

```python
import pandas as pd

rain_ar=[]
all_ar=[]
rain_dic={}
Tcnt = 0
Xcnt = 0
Rcnt = 0
T = 0.3 
X = None
for d in taipei_data:
    all = d["weatherElements"]
    rain = d["weatherElements"]["Precipitation"]
    datetime= d["DateTime"]
    if rain == "X":
        Xcnt += 1
        rain = X
    elif rain == "T":
        Tcnt += 1
        rain = T
    else:
        rain = float(rain)
        if rain < 0.0:
            raise ValueError("rain is negative")
        if rain > 0.0:
            Rcnt+=1
        if rain in rain_dic:
            rain_dic[rain]+=1
        else:
            rain_dic[rain]=1
    if "24:00:00" in datetime:
        datetime = datetime.split('T')[0]
        datetime = pd.to_datetime(datetime)+pd.Timedelta(days=1)
    else:
        datetime = pd.to_datetime(datetime).tz_localize(None)
    rain_ar.append({"datetime":datetime,"rain":rain})
    all= {"datetime": datetime,**all}
    all_ar.append(all)
print(f"Total: {len(taipei_data)} (dup)")
print(Tcnt)
print(Rcnt)
print(Xcnt)
print(rain_dic)
```
